```python
#feature mapping
def map_feature(X1,X2):
  X1=np.atleast_1d(X1)
  X2=np.atleast_1d(X2)
  degree=6
  out=[]
  for i in range(1,degree+1):
    for j in range(i+1):
      out.append((X1**(i-j) * (X2**j)))
  return np.stack(out, axis=1)
```

```python
def compute_cost(X, y, w, b, lambda_=1):
    """
    Computes the cost function for logistic regression with regularization.

    Args:
        X (numpy.ndarray): The feature matrix.
        y (numpy.ndarray): The target values.
        w (numpy.ndarray): The weight vector.
        b (float): The bias term.
        lambda_ (float, optional): The regularization parameter. Defaults to 1.

    Returns:
        float: The computed cost.
    """
    m = X.shape[0]  # number of training examples

    # Compute the predictions
    z_wb = np.dot(X, w) + b
    predictions = sigmoid(z_wb)  # Apply sigmoid activation to get predictions

    # Compute the cost without regularization
    epsilon = 1e-10  # Small value to avoid division by zero in log calculations
    cost = (-1 / m) * np.sum(y * np.log(predictions + epsilon) + (1 - y) * np.log(1 - predictions + epsilon))

    # Compute the regularization term
    reg_cost = (lambda_ / (2 * m)) * np.sum(w ** 2)

    # Total cost
    total_cost = cost + reg_cost

    return total_cost
```

```python
#compute regularised cost
def compute_cost_reg(X, y, w, b, lambda_ = 1):
    """
    Computes the cost function for logistic regression with regularization.

    Args:
        X (numpy.ndarray): The feature matrix.
        y (numpy.ndarray): The target values.
        w (numpy.ndarray): The weight vector.
        b (float): The bias term.
        lambda_ (float, optional): The regularization parameter. Defaults to 1.

    Returns:
        float: The computed cost.
    """
    m = X.shape[0]  # number of training examples

    # Compute the predictions
    z_wb = np.dot(X, w) + b
    predictions = sigmoid(z_wb)  # Apply sigmoid activation to get predictions

    # Compute the cost without regularization
    epsilon = 1e-10  # Small value to avoid division by zero in log calculations
    cost = (-1 / m) * np.sum(y * np.log(predictions + epsilon) + (1 - y) * np.log(1 - predictions + epsilon))

    # Compute the regularization term
    reg_cost = (lambda_ / (2 * m)) * np.sum(w ** 2)

    # Total cost
    total_cost = cost + reg_cost

    return total_cost
```

```python
def compute_gradient(X, y, w, b, lambda_=None):
    m, n = X.shape  # (number of records, number of features)
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):
        z_wb = np.dot(X[i], w) + b
        err = sigmoid(z_wb) - y[i]  # Calculate error for logistic regression
        for j in range(n):
            dj_dw[j] += err * X[i, j]  # Update dj_dw
        dj_db += err  # Update dj_db

    dj_dw /= m  # Divide dj_dw by m
    dj_db /= m  # Divide dj_db by m (number of samples) to get the average gradient

    if lambda_ is not None:
        dj_dw += (lambda_ / m) * w  # Regularization term for dj_dw

    return dj_db, dj_dw
```
